```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error , mean_squared_error
```

```python
weather=pd.read_csv(r"C:\Users\Mega Store\Desktop\python mia\weatherHistory.csv")
```

```python
plt.figure(figsize=(10,10))
plt.scatter(weather["Humidity"],weather["Temperature (C)"],alpha=.5)
plt.xlabel("Humidity")
plt.ylabel("Temperature")
plt.xlim(-.1,1.1)
plt.ylim(-30,45)
plt.title("Temperature vs Humidity")
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(weather["Apparent Temperature (C)"],weather["Temperature (C)"],alpha=.5)
plt.xlabel("Apparent Temperature (C)")
plt.ylabel("Temperature")
plt.title("Temperature vs Apparent Temperature (C)")
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(weather["Wind Speed (km/h)"],weather["Temperature (C)"],alpha=.5)
plt.xlabel("Wind Speed (km/h)")
plt.ylabel("Temperature")
plt.title("Temperature vs Wind Speed (km/h)")
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(weather["Wind Bearing (degrees)"],weather["Temperature (C)"],alpha=.5)
plt.xlabel("Wind Bearing (degrees)")
plt.ylabel("Temperature")
plt.title("Temperature vs Wind Bearing (degrees)")
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(weather["Visibility (km)"],weather["Temperature (C)"],alpha=.5)
plt.xlabel("Visibility (km)")
plt.ylabel("Temperature")
plt.title("Temperature vs Visibility (km)")
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(weather["Pressure (millibars)"],weather["Temperature (C)"],alpha=.5)
plt.xlabel("Pressure (millibars)")
plt.ylabel("Temperature")
plt.title("Temperature vs Pressure (millibars)")
plt.show()
```

```python
x=weather[["Humidity","Apparent Temperature (C)","Wind Speed (km/h)","Wind Bearing (degrees)","Visibility (km)","Pressure (millibars)"]]
y=weather["Temperature (C)"]
x_train , x_test , y_train , y_test =train_test_split(x,y,test_size=.2,random_state=42)
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
print(f"Mean absolute error: {mae}")
print(f"Mean sqared error: {mse}")
```

```python
plt.figure(figsize=(10,8))
plt.scatter(y_test,y_pred,alpha=.5)
plt.xlabel("Acual Temperature")
plt.ylabel("Predicted Temperature")
plt.title("Acual Temperature vs Predicted Temperature")
plt.show()
```
