```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Mega Store\Desktop\python mia\Weather_Dataset.csv")
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
df.sort_values(by='Formatted Date', inplace=True)
df = df[(df['Temperature (C)'] >= -50) & (df['Temperature (C)'] <= 50) & (df["Apparent Temperature (C)"] >= -50) & (df["Apparent Temperature (C)"] <= 50) & (df["Pressure (millibars)"]>1000)]
df = df[(df["Wind Bearing (degrees)"]>=20)& (df["Loud Cover"]>=0)]
plt.figure(figsize=(12, 6))
plt.plot(df['Formatted Date'], df['Temperature (C)'], linestyle='-')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.title('Temperature Over Time')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(df["Temperature (C)"],bins=20, edgecolor="black")
plt.xlabel("Temperature")
plt.ylabel("Frequency")
plt.title("Temperature Distribution")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Temperature (C)'], df['Humidity'], alpha=.8)
plt.ylim(.2,1.3)
plt.xlim(-10,40)
plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.title('Temperature vs Humidity')
plt.grid(True)
plt.show()
```

```python
removed_coloumns=["Formatted Date","Summary","Precip Type","Daily Summary","Loud Cover"]
for c in removed_coloumns:
    if c in df.columns:
        df.drop(columns=[c],inplace=True)
my_matrix=df.corr().round(2)
plt.figure(figsize=(8,8))
sns.heatmap(my_matrix,cmap="Greens",vmin=0,vmax=1,annot=True,fmt=".2f",square=True)
plt.show()
```
